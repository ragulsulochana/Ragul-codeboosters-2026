Now, we'll initialize the embedding model and the `chromadb` client and collection.

```python
import chromadb
from sentence_transformers import SentenceTransformer

# Initialize an embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Initialize chromadb client and collection
client = chromadb.PersistentClient(path='./chroma_db') # Store the database locally
collection_name = 'college_notes_collection'

try:
    collection = client.get_collection(name=collection_name)
    print(f"Collection '{collection_name}' already exists. Using existing collection.")
except:
    collection = client.create_collection(name=collection_name)
    print(f"Collection '{collection_name}' created.")
```

Next, we will generate embeddings for the 'content' column of your DataFrame and add the notes to the `chromadb` collection.

```python
# Prepare data for chromadb
documents = df['content'].tolist()
metadatas = df[['subject', 'topic']].to_dict(orient='records')
ids = df['note_id'].tolist()

# Generate embeddings for the content
embeddings = model.encode(documents).tolist()

# Add data to the chromadb collection
collection.add(
    embeddings=embeddings,
    documents=documents,
    metadatas=metadatas,
    ids=ids
)

print(f"Successfully added {len(ids)} notes to the '{collection_name}' collection.")
print(f"Number of items in the collection: {collection.count()}")
```

Now that our notes are indexed in ChromaDB, we can accept student questions as input and retrieve relevant notes.

```python
def query_notes(question, k=3):
    """
    Queries the chromadb collection for notes relevant to the given question.

    Args:
        question (str): The student's question.
        k (int): The number of top relevant notes to retrieve.

    Returns:
        dict: The query results from chromadb.
    """
    # Embed the question
    question_embedding = model.encode([question]).tolist()

    # Query the chromadb collection
    results = collection.query(
        query_embeddings=question_embedding,
        n_results=k,
        include=['documents', 'metadatas', 'distances']
    )
    return results

# Example usage:
student_question = input("Enter a student question: ")

if student_question:
    retrieved_notes = query_notes(student_question)
    print("\n--- Retrieved Notes ---")
    for i, (doc, meta, dist) in enumerate(zip(retrieved_notes['documents'][0], retrieved_notes['metadatas'][0], retrieved_notes['distances'][0])):
        print(f"Note {i+1} (Distance: {dist:.4f}):")
        print(f"  Subject: {meta['subject']}")
        print(f"  Topic: {meta['topic']}")
        print(f"  Content: {doc[:200]}...") # Display first 200 chars
        print("---")
else:
    print("No question entered. Please try again.")
```
