```python
df = raw_df.copy()
print(f'working copy created: (df.shape)')
print('raw_df is untouched - we can always reset ')
```

```python
#FIX 1

print("Before fixing nulls:", df.isnull().sum().sum())

df['customer_name'].fillna("No Name", inplace=True)
median_quantity = df['quantity'].median()
df['quantity'].fillna(median_quantity, inplace=True)
print("Filled missing values in the qnantity with median", median_quantity)

df['category'].fillna("No Category", inplace=True)

df['product'].fillna("No Product", inplace=True)
print("After fixing nulls:", df.isnull().sum().sum())
```

```python
#FIX 2 REMOVING DUPLICATE VALUES

print("Before Duplication: ", len(df))
print("Duplicate rows: ", df.duplicated().sum())
print("\nDuplicate rows:")
print(df[df.duplicated(keep=False)][['order_id', 'customer_name', 'product', 'category', 'quantity', 'unit_price', 'order_date', 'city', 'sales_rep']])

df.drop_duplicates(inplace=True)

print("\nAfter duplication:", )
```

```python
print("Sample dates before parsing: ")
print(df['order_date'].head(5).tolist())

df['order_date'] = pd.to_datetime(
    df['order_date'],
    dayfirst=False,
    errors='coerce'
)
nat_count = df['order_date'].isnull().sum()
print(f"\nUnparseable dates (NaT): {nat_count}")

# Fixed: removed parentheses from .year and .month
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.strftime('%B')

print('\nSample Date after parsing:')
print(df[['order_date', 'year', 'month', 'month_name']].head(5))
```
