```python
import pandas as pd
```

```python
df = pd.read_csv("/content/student_performance.csv")
```

```python
print(f"Total Students: {len(df)}")
print(f"Total Columns: {df.shape[1]}")
print(f"Total Departments: {df['department'].nunique()}")
```

```python
dept = df.groupby('department').size()
print(dept)
```

```python
print("highest maths score : ",df.max()['math_score'])
print("lowest maths score : ",df.min()['math_score'])
print("Average maths score : ",df.mean(numeric_only=True)['math_score'])
print("Average programming score : ",df.mean(numeric_only=True)['programming_score'])
print("Average Attendance : ",df.mean(numeric_only=True)['attendance_percentage'])
```

```python
print("Gender wise Average score")
print("-"*40)
total_score = df['math_score']+df['programming_score']+df['science_score']+df['english_score']
df['total_score'] = total_score
gender_score = df.groupby('gender')['total_score'].mean().round(2)
print(gender_score)
```

```python
print("Top 5 Performers:")
print("-"*40)

top_performers = df.sort_values(by='total_score', ascending=False).head(5)

for index, row in top_performers.iterrows():
    print(f"Name: {row['name']}, Total Score: {row['total_score']}")
```

```python
print("Students of Lower Attendance")
print("="*50)
low_at = df[df['attendance_percentage']<75][['name','department','attendance_percentage']]
if len(low_at)>0:
  print(low_at.to_string(index=False))
else:
  print("All the Students are Greater than 75 %")
```
