```python
cs_students = df [df['department'] == 'Computer Science']
print(f"number of computer science students: {len(cs_students)}")

print("computer science students:")
print(cs_students[['name','math_score','programming_score','attendance_percentage']])
```

```python
cs_database_high_attendance = df[(df['department'] == 'Computer Science') & (df['attendance_percentage'] > 90)]
print("Computer Science students with attendance score above 90%:")
print(cs_database_high_attendance[['name', 'attendance_percentage']])
```

```python
df['mean_score'] = df[['math_score', 'science_score', 'english_score', 'programming_score']].mean(axis=1)

top_performers = df[df['mean_score'] > 70]

print("Top Performers based on mean subject marks:")
print(top_performers[['name', 'department', 'mean_score']])
```

```python
cs_students = df[df['department'] == 'Computer Science']
mech_students = df[df['department'] == 'Mechanical']
elec_students = df[df['department'] == 'Electronics']
civil_students = df[df['department'] == 'Civil']

print(f"Average math score of CS Dept: {cs_students['math_score'].mean(axis=0)}")
print(f"Average math score of Mech Dept: {mech_students['math_score'].mean(axis=0)}")
print(f"Average math score of Electronics Dept: {elec_students['math_score'].mean(axis=0)}")
print(f"Average math score of Civil Dept: {civil_students['math_score'].mean(axis=0)}")
```

```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
dept_avg = df.groupby('department')['total_score'].mean().round()
print("Average total score of each department:")
print(dept_avg)
```

```python
print(dept_avg.sort_values(ascending=False))
```
