```python
# Step 2: Importing Libraries

import pandas as pd
import chromadb

from sentence_transformers import SentenceTransformer

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Step 3: Creating Dataset

internships = [
    {
        "id": 1,
        "title": "AI Research Intern",
        "description": "Work on machine learning models, deep learning, and natural language processing."
    },
    {
        "id": 2,
        "title": "Frontend Developer Intern",
        "description": "Build responsive web applications using React and JavaScript."
    },
    {
        "id": 3,
        "title": "Cybersecurity Intern",
        "description": "Perform vulnerability assessments and security audits."
    },
    {
        "id": 4,
        "title": "Data Analyst Intern",
        "description": "Analyze business data using Python, SQL, and visualization tools."
    },
    {
        "id": 5,
        "title": "Cloud Computing Intern",
        "description": "Deploy scalable applications using AWS and Docker."
    },
    {
        "id": 6,
        "title": "Computer Vision Intern",
        "description": "Develop image classification and object detection systems."
    },
    {
        "id": 7,
        "title": "DevOps Intern",
        "description": "Manage CI/CD pipelines and cloud infrastructure."
    },
    {
        "id": 8,
        "title": "Backend Developer Intern",
        "description": "Develop REST APIs using Node.js and MongoDB."
    },
    {
        "id": 9,
        "title": "Robotics Intern",
        "description": "Program autonomous robots using sensors and AI algorithms."
    },
    {
        "id": 10,
        "title": "Generative AI Intern",
        "description": "Build applications using LLMs, embeddings, vector databases, and RAG."
    }
]
```

```python
# Step 4: Converting to CSV

df = pd.DataFrame(internships)

df.to_csv("internships.csv", index=False)

print(df)
```

```python
# Step 5: Creating Embeddings

model = SentenceTransformer('all-MiniLM-L6-v2')

documents = df["description"].tolist()

embeddings = model.encode(documents)

print("Embedding Shape:")
print(embeddings.shape)
```

```python
# Step 6: Cosine Similarity

query = "I want to work with LLMs and AI chatbots"

query_embedding = model.encode([query])

scores = cosine_similarity(
    query_embedding,
    embeddings
)

scores

for i, score in enumerate(scores[0]):
    print(df.iloc[i]["title"], "->", round(score,4))
```

```python
# Step 7: Semantic Search

def semantic_search(query, top_k=3):

    query_embedding = model.encode([query])

    similarities = cosine_similarity(
        query_embedding,
        embeddings
    )[0]

    top_indices = similarities.argsort()[-top_k:][::-1]

    print(f"\nQuery: {query}\n")

    for idx in top_indices:
        print("Title:", df.iloc[idx]["title"])
        print("Score:", round(similarities[idx],4))
        print("Description:", df.iloc[idx]["description"])
        print()
```

```python
semantic_search("AI chatbot development")

semantic_search("Finding security vulnerabilities")

semantic_search("Cloud deployment and containers")
```

```python
# Step 8: ChromaDB

client = chromadb.Client()

collection = client.create_collection(
    name="internships"
)

# Inserting Data into ChromaDB
for i in range(len(df)):

    collection.add(
        ids=[str(df.iloc[i]["id"])],

        documents=[
            df.iloc[i]["description"]
        ],

        metadatas=[
            {
                "title": df.iloc[i]["title"]
            }
        ]
    )

# Querying Data from ChromaDB
results = collection.query(
    query_texts=[
        "Large Language Models and RAG"
    ],
    n_results=3
)

results

# Displaying Resulta
for i in range(len(results["documents"][0])):

    print(
        results["metadatas"][0][i]["title"]
    )

    print(
        results["documents"][0][i]
    )

    print()
```
