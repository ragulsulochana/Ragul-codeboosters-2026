```python
model = SentenceTransformer('all-MiniLM-L6-v2')

print("Model loaded sucessfully")
print(f"Model produces vectors of size: {model.get_sentence_embedding_dimension()} dimensions")
```

```python
sentence = "ETL is used to clean and transform data"

embedding = model.encode(sentence)

print(f"Input Sentence: {sentence}")
print(f"Embedding Type: {type(embedding)}")
print(f"Embedding Shape: {embedding.shape}")
print(f"First 10 Embedding Numbers: {embedding[:10]}")
print(f"Min Embedding Value: {embedding.min():.2f}")
print(f"Max Embedding Value: {embedding.max():.2f}")
# print(embedding)
```

```python
chroma_client = chromadb.Client()

collection = chroma_client.get_or_create_collection(name="demo_notes")

documents = [
    "Python is a popular programming language",
    "Quality Education: Providing outcome- and value-based engineering education.",
    "Education is the systematic process of acquiring knowledge, skills, values, and habits through instruction, training, or research",
    "Learning to Know: Developing critical thinking and mastering learning tools.",
    "Early Childhood: Foundational learning focusing on social and basic motor skills."
]

ids = [f"doc_{i}" for i in range(len(documents))]

collection.add(
    documents=documents,
    ids=ids
)

print("Collection initialized sucessfully")
print(f"Collection Name: {collection.name}")
print(f"Collection Count: {collection.count()}")
```

```python
sample_docs = [
    "ETL is used to clean and transform data",
    "A vehicle is a mode of transportation",
    "Cars and trucks are popular automobiles",
    "SQL is used to query databases",
    "Machine learning trains models on data",
]

sample_ids = ["doc001", "doc002", "doc003", "doc004", "doc005"]

sample_metadata = [
    {"subject": "Data Engineering", "topic": "ETL"},
    {"subject": "Transportation", "topic": "Vehicle"},
    {"subject": "Automobile", "topic": "Popular"},
    {"subject": "Database", "topic": "SQL"},
    {"subject": "Machine Learning", "topic": "Training"}
]

collection.add(
    documents=sample_docs,
    ids=sample_ids,
    metadatas=sample_metadata
)

print("Document Added Sucesssfully")
print(f"Total Documents Count: {collection.count()}")
```

```python
query = "How do I clean and prepare data?"

results = collection.query(
    query_texts=[query],
    n_results=3
)

print(f"Query: {query}")
print(list(results.keys()))
print()
```

```python
print(f"Query: '{query}'")
print("="*60)

matched_docs = results['documents'][0]
matched_ids = results['ids'][0]
matched_meta = results['metadatas'][0]
matched_dists = results['distances'][0]

for rank, (doc, doc_id, dist, meta) in enumerate(zip(matched_docs, matched_ids, matched_dists, matched_meta)):
  print(f"Rank {rank+1}:")
  print(f"  Document: {doc}")
  print(f"  Document ID: {doc_id}")
  print(f"  Distance: {dist}")
  print(f"  Metadata: {meta}")
```

```python
query = "How do computers learn from examples?"

filtered_results = collection.query(
    query_texts=[query],
    n_results=3,
    where={"subject": "Machine Learning"}
)

print(f"Filtered Query: {query}")
matched_docs = filtered_results['documents'][0]
matched_ids = filtered_results['ids'][0]
matched_meta = filtered_results['metadatas'][0]
matched_dists = filtered_results['distances'][0]

for rank, (doc, doc_id, dist, meta) in enumerate(zip(matched_docs, matched_ids, matched_dists, matched_meta)):
  print(f"Rank {rank+1}:")
  print(f"  Document: {doc}")
  print(f"  Document ID: {doc_id}")
  print(f"  Distance: {dist}")
  print(f"  Metadata: {meta}")
```
